```python
import pandas as pd

df = pd.read_csv('/content/drive/My Drive/finalcorpus.csv', ';', encoding= 'unicode_escape')

df['text'].head()
```

```python
df['label'].value_counts()
```

```python
def clean_text(text):
    text = text.lower()

    text = " ".join(text.split())
    return text


df['clean']= df['text'].map(clean_text)

df
```

```python
df['text_clean'] = df['clean'].str.replace(r'[^\w\s]+', '')
df
```

```python
df1 = pd.read_csv(r'C:\Users\mtsil\OneDrive\Documents\cleaneddoc.csv', ';', encoding= 'unicode_escape') #call newdf as df1
df1.head()
df1['label'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df1['cleaned'],df1['label'], test_size=0.2, random_state=42)
```

```python
from sklearn.feature_extraction.text import CountVectorizer


cv = CountVectorizer(max_df=200, min_df=20)

# convert training data to bag of words
X_train_cv = cv.fit_transform(X_train)
X_test_cv = cv.transform(X_test)
```

```python
from sklearn.metrics import f1_score

from sklearn.naive_bayes import BernoulliNB
model = BernoulliNB()
model.fit(X_train_cv, y_train)
y_pred = model.predict(X_test_cv)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
